```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv("epa-sea-level.csv")
```

```python
def create_scatter_plot():
    plt.figure(figsize=(12, 6))
    plt.scatter(df["Year"], df["CSIRO Adjusted Sea Level"])
    plt.xlabel("Year")
    plt.ylabel("Sea Level (inches)")
    plt.title("Rise in Sea Level")
```

```python
def get_linregress(x, y):
    slope, intercept, _, _, _ = linregress(x, y)
    return slope, intercept
```

```python
def plot_line_of_best_fit():
    slope, intercept = get_linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
    plt.plot(df["Year"], intercept + slope * df["Year"], "r", label="Line of Best Fit")
    plt.plot(2050, intercept + slope * 2050, "g*", markersize=10, label="Prediction for 2050")
```

```python
def plot_line_of_best_fit_from_2000():
    recent_df = df[df["Year"] >= 2000]
    slope, intercept = get_linregress(recent_df["Year"], recent_df["CSIRO Adjusted Sea Level"])
    plt.plot(recent_df["Year"], intercept + slope * recent_df["Year"], "b", label="Line of Best Fit (since 2000)")
    plt.plot(2050, intercept + slope * 2050, "g*", markersize=10, label="Prediction for 2050")
```

```python
create_scatter_plot()
plot_line_of_best_fit()
plot_line_of_best_fit_from_2000()
plt.legend()
plt.show()
```
